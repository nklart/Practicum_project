## Описание проекта

HR-аналитики компании помогают бизнесу оптимизировать управление персоналом: бизнес предоставляет данные, а аналитики предлагают, как избежать финансовых потерь и оттока сотрудников. В этом HR-аналитикам пригодится машинное обучение, с помощью которого получится быстрее и точнее отвечать на вопросы бизнеса.


Компания предоставила данные с характеристиками сотрудников компании. Среди них — уровень удовлетворённости сотрудника работой в компании. Эту информацию получили из форм обратной связи: сотрудники заполняют тест-опросник, и по его результатам рассчитывается доля их удовлетворённости от 0 до 1, где 0 — совершенно неудовлетворён, 1 — полностью удовлетворён.


Собирать данные такими опросниками не так легко: компания большая, и всех сотрудников надо сначала оповестить об опросе, а затем проследить, что все его прошли.


У нас будет несколько задач. Первая — построить модель, которая сможет предсказать уровень удовлетворённости сотрудника на основе данных заказчика.


Почему бизнесу это важно: удовлетворённость работой напрямую влияет на отток сотрудников. А предсказание оттока — одна из важнейших задач HR-аналитиков. Внезапные увольнения несут в себе риски для компании, особенно если уходит важный сотрудник.


Наша вторая задача — построить модель, которая сможет на основе данных заказчика предсказать то, что сотрудник уволится из компании.

````python
# создадим функцию для построения диаграмм категориальных признаков
def plot_cat_bar(df, labels):
    
    # Получим список категориальных признаков
    cat_columns = df.select_dtypes(include='object').columns.tolist()
    
    # Задаем область рисования
    fig, axs = plt.subplots(
        ncols=1,
        nrows=len(cat_columns),
        figsize=(10, len(cat_columns) * 5)
    )
    
    def plot_count_once(column_name, label, ax):
        sns.countplot(
            data=df,
            x=column_name,
            color='skyblue',
            ax=ax
        )
        
        # Подписываем ось Y
        ax.set_ylabel('Количество сотрудников')

        # Подписываем ось X
        ax.set_xlabel(label)

        # Устанавливаем в качестве названия графика название столбца
        title = f'Feature name: {column_name}'
        ax.set_title(title, fontsize=14, fontweight='bold')

    # Если в датасете один признак для вывода, рисуем график 1 раз
    if len(cat_columns) == 1:
        plot_count_once(column_name=cat_columns[0], label=labels[0], ax=axs)
        
    # Если признаков больше 1, рисуем график в цикле для каждого признака
    else:          
        for i, ax in enumerate(axs):
            plot_count_once(column_name=cat_columns[i], label=labels[i], ax=ax)

    # Устанавливаем расстояния между графиками
    plt.subplots_adjust(hspace=0.5)
````

````python
# создадим функцию для построения гистограмм количественных признаков
def plot_num_hist(df, labels):
    
    # получим список количественных признаков
    num_columns = df.select_dtypes(include='number').columns.tolist()
    
    # задаем область рисования
    fig, axs = plt.subplots(
        ncols=1,
        nrows=len(num_columns),
        figsize=(10, len(num_columns) * 5)
    )
    
    def plot_hist_once(column_name, label, ax):
        sns.histplot(
            data=df[column_name],
            color='skyblue',
            ax=ax
        )
        
        # подписываем ось Y
        ax.set_ylabel('Количество сотрудников в %')

        # подписываем ось X
        ax.set_xlabel(label)

        # устанавливаем в качестве названия графика название столбца
        title = f'Feature name: {column_name}'
        ax.set_title(title, fontsize=14, fontweight='bold')
        

    # если в датасете один признак для вывода, рисуем гистограмму 1 раз
    if len(num_columns) == 1:
        plot_hist_once(column_name=num_columns[0], label=labels[0], ax=axs)
        
        
    # если признаков больше 1, рисуем гистограмму в цикле для каждого признака
    else:          
        for i, ax in enumerate(axs):
            plot_hist_once(column_name=num_columns[i], label=labels[i], ax=ax)
            
    # Устанавливаем расстояния между графиками
    plt.subplots_adjust(hspace=0.5)
````

````python
# создадим функцию для построения матрицы корреляции
def plot_correlation_matrix(df):
    # Создание корреляционной матрицы
    corr_matrix = df.corr()
    
    # Установка размеров фигуры
    plt.figure(figsize=(12, 10))
    
    # Построение тепловой карты с помощью seaborn
    sns.heatmap(
        corr_matrix,
        annot=True,            # Показывать коэффициенты корреляции
        fmt=".2f",             # Форматирование значений
        cmap='Blues',          # Цветовая палитра
        cbar=True,             # Отображение цветового бара
        square=True            # Квадратные ячейки
    )
    
    # Установка заголовка
    plt.title('Матрица корреляции признаков', fontweight='bold', fontsize=16)
    
    # Показать график
    plt.show()
````

````python
# создадим функцию для построния боксплотов
def plot_boxplots(data, title='Боксплоты для числовых столбцов'):
    
    numeric_cols = data.select_dtypes(include=[float, int]).columns  # Определение числовых столбцов
    num_columns = len(numeric_cols)
    
    if num_columns == 0:
        print("В DataFrame нет числовых столбцов.")
        return
    
    plt.figure(figsize=(7, 5 * num_columns))  # Установка размера фигуры
    
    for i, column in enumerate(numeric_cols):
        plt.subplot(num_columns, 1, i + 1)  # Создание подграфика для каждого столбца
        sns.boxplot(y=data[column])  # Построение боксплота
        plt.title(f'Боксплот для "{column}"', fontsize=14)
        plt.xlabel(column)
    
    plt.tight_layout()  # Уплотнение подграфиков
    plt.suptitle(title, fontsize=20, fontweight='bold', y=1.05)  # Заголовок для всей фигуры
    plt.show()  # Отображение графиков
````

````python
# создаем списки с названиями признаков
ohe_columns = ['dept', 'last_year_promo', 'last_year_violations']
ord_columns = ['level', 'workload']
num_columns = ['employment_years', 'supervisor_evaluation', 'salary']
````

````python
# создаем конвейер для обработки данных с помощью One-Hot Encoding
ohe_pipeline = Pipeline(
    [
        (
            'simpleImputer_ohe',  # первый шаг
            SimpleImputer(missing_values=np.nan, strategy='most_frequent')  
        ),
        (
            'ohe',  # второй шаг
            OneHotEncoder(drop='first', sparse_output=False, handle_unknown='ignore')  # применяем OHE
        )
    ]
)
ohe_pipeline
````

````python
# определение конвейера
ord_pipeline = Pipeline([
    (
        'simpleImputer_before_ord',
        SimpleImputer(missing_values=np.nan, strategy='most_frequent')  
    ),
    # кодируем категориальные признаки
    (
        'ord',
        OrdinalEncoder(  # применяем OrdinalEncoder для кодирования категориальных признаков
            categories=[['junior', 'middle', 'senior'],
                        ['low', 'medium', 'high']],
            handle_unknown='use_encoded_value', unknown_value=np.nan
    )
    ),
    (
        'simpleImputer_after_ord',
        SimpleImputer(missing_values=np.nan, strategy='most_frequent') 
    )
])
ord_pipeline
````

````python
# объединим кодирование, масштабирование и подготовку данных в один пайплайн
data_preprocessor = ColumnTransformer(
    [
        ('ohe', ohe_pipeline, ohe_columns),  # применяем ohe_pipeline к ohe_columns
        ('ord', ord_pipeline, ord_columns),  # применяем ord_pipeline к ord_columns
        ('num', MinMaxScaler(), num_columns)
    ],
    remainder='passthrough'
)
data_preprocessor
````

````python
# создаём итоговый пайплайн: подготовка данных и модель
general_pipeline = Pipeline([
    ('preprocessor', data_preprocessor),
    ('models', LinearRegression(n_jobs=-1)) 
])
general_pipeline
````

````python
# создадим функцию
def pipeline(ohe_columns, ord_columns, num_columns):
    global general_pipeline
    ohe_pipeline = Pipeline(
        [
            (
                'simpleImputer_ohe',  # первый шаг
                SimpleImputer(missing_values=np.nan, strategy='most_frequent')  
            ),
            (
                'ohe',  # Название второго шага в конвейере
                OneHotEncoder(drop='first', sparse_output=False, handle_unknown='ignore')  # применяем OHE
            )
        ]
    )

    ord_pipeline = Pipeline([
        (
            'simpleImputer_before_ord',
            SimpleImputer(missing_values=np.nan, strategy='most_frequent')  
        ),
        # Кодируем категориальные признаки
        (
            'ord',
            OrdinalEncoder(  # Применяем OrdinalEncoder для кодирования категориальных признаков
                categories=[['junior', 'middle', 'senior'],
                            ['low', 'medium', 'high']],
                handle_unknown='use_encoded_value', unknown_value=np.nan
            )
        ),
        (
            'simpleImputer_after_ord',
            SimpleImputer(missing_values=np.nan, strategy='most_frequent') 
        )
    ])  

    data_preprocessor = ColumnTransformer(
        [
            ('ohe', ohe_pipeline, ohe_columns),  # Применяем ohe_pipeline к ohe_columns
            ('ord', ord_pipeline, ord_columns),  # Применяем ord_pipeline к ord_columns
            ('num', MinMaxScaler(), num_columns)
        ],
        remainder='passthrough'
    )

    general_pipeline = Pipeline([
        ('preprocessor', data_preprocessor),
        ('models', LinearRegression(n_jobs=-1)) 
    ])
````

## Шаг 5. Обучение моделей

### 5.1 Создадим функцию для выбора лучшей модели по метрик SMAPE

````python
def smape(y_true, y_pred):
    return 100/len(y_true) * np.sum(2 * np.abs(y_pred - y_true) / (np.abs(y_true) + np.abs(y_pred)))
   
smape_scorer = make_scorer(score_func=smape, greater_is_better=False)
````

### 5.2 Перебор параметров

````python
param_grid = [
    {    # линейная регрессия
        'models': [LinearRegression()],
        'preprocessor__num': [StandardScaler(), MinMaxScaler(), 'passthrough']   
    },
    {    # дерево решений
        'models': [DecisionTreeRegressor(random_state=RANDOM_STATE)],
        'models__max_depth': range(5, 15),
        'models__max_features': range(5, 15),
        'preprocessor__num': [StandardScaler(), MinMaxScaler(), 'passthrough']  
    }
]
````

### 5.3 Поиск лучшей модели

````python
grid_search = GridSearchCV(
    general_pipeline,
    param_grid,
    n_jobs=-1,
    cv=5,
    scoring=smape_scorer
)
grid_search.fit(X_train, y_train)
````

````python
best_model = grid_search.best_estimator_
print('Лучшая модель:\n\n', grid_search.best_estimator_)
print ('Метрика лучшей модели:', grid_search.best_score_*(-1))
````

````python
# значения метрики SMAPE на тестовой выборке
y_test_pred_job = grid_search.best_estimator_.predict(X_test)
print(f'Метрика SMAPE на тестовой выборке: {smape(y_test, y_test_pred_job)}')
````
